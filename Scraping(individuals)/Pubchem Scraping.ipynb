```python
import pandas as pd
data = pd.read_excel("Property_and_functionalities.xlsx")
```

```python
from tqdm import tqdm
```

```python
import os
import time
from pubchempy import Compound, get_compounds

ids = []
for name in tqdm(data['Chemical Name']):
    for compound in get_compounds(name, 'name'):
        #print(compound)
        if name+'.json' not in os.listdir('Pubchem_jsons'):
            time.sleep(2)
            ids.append(compound.cid)
            try:
                url = "https://pubchem.ncbi.nlm.nih.gov/rest/pug_view/data/compound/" + str(compound.cid)\
        + "/JSON/?response_type=save&response_basename=compound_CID_" + str(compound.cid)

                r = requests.get(url, allow_redirects=True)

                open('Pubchem_jsons/' + name + '.json', 'wb').write(r.content)
            except:
                #print(compound)
                continue
```

```python
import json
all_property_dicts = []
all_uses = []
for chem_name in data['Chemical Name']:
    try:
        x = json.load(open('Pubchem_jsons/' + chem_name + '.json', 'rb'))
        for rec in x['Record']['Section']:
            #print("Out:" ,rec['TOCHeading'])
            all_properties = {}
            if rec['TOCHeading'] == 'Chemical and Physical Properties':
                uses = rec
                
                for section in uses['Section']:
                    #print("In:",section['TOCHeading'])
                    if section['TOCHeading'] == 'Computed Properties':
                        for s in section['Section']:
                            #print(s['TOCHeading'])
                            #print(s['Information'])
                            if 'StringWithMarkup' in s['Information'][0]['Value'].keys():
                                if 'Unit' in s['Information'][0]['Value'].keys():
                                    all_properties[s['TOCHeading']] = s['Information'][0]['Value']['StringWithMarkup'][0]['String'][0] +\
                                                      s['Information'][0]['Value']['Unit']
                                else:
                                    all_properties[s['TOCHeading']] = s['Information'][0]['Value']['StringWithMarkup'][0]['String'][0]
                            elif 'Number' in s['Information'][0]['Value'].keys():
                                all_properties[s['TOCHeading']] = (s['Information'][0]['Value']['Number'][0])
                            all_properties['Name'] = chem_name
                            #print(all_properties)
                       
            if rec['TOCHeading'] == 'Names and Identifiers':
                #re
                for section in rec['Section']:
                    #print("Two:",section['TOCHeading'])
                    if section['TOCHeading'] == 'Molecular Formula':
                        #print(section)
                        if 'StringWithMarkup' in section['Information'][0]['Value'].keys():
                            #print(section['Information'][0]['Value']['StringWithMarkup'][0]['String'][0])
                            all_properties['Molecular Formula'] = section['Information'][0]['Value']['StringWithMarkup'][0]['String']
                              #             else:
            #                 all_properties.append(s['Information'][0]['Value']['Number'][0])
                        #all_properties['Name'] = chem_name
                        #print(all_properties)
            all_property_dicts.append(all_properties)
                
        #all_industry_uses = [x['String'] for x in industry_uses['Information'][0]['Value']['StringWithMarkup']]
        #all_uses.append(all_industry_uses)
    except Exception as e:
        print(e)
        #print(chem_name)
```
